# Visualization Examples

This is an example of how to use the generated visualization module.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams.update({'figure.figsize': (12, 6), 'figure.dpi': 150})
```

## 1. Load Data

```python
df = pd.read_parquet('outputs/processed/ais_features.parquet')
print(f'Records: {len(df):,}')
print(f'Columns: {list(df.columns)}')
```

## 2. Spatial Visualization (Folium)

```python
try:
    from src.visualization.spatial_viz import SpatialVisualizer
    
    viz = SpatialVisualizer(output_dir='outputs/figures/spatial')
    viz.plot_density_heatmap(df.sample(min(10000, len(df))))
    viz.plot_ais_coverage(df)
    print('Spatial visualizations saved.')
except Exception as e:
    print(f'Spatial viz error: {e}')
```

## 3. Temporal Visualization

```python
from src.visualization.temporal_viz import TemporalVisualizer

viz = TemporalVisualizer(output_dir='outputs/figures/temporal')
viz.plot_daily_traffic(df)
viz.plot_dark_ship_ratio(df)
viz.plot_mean_rot(df)
print('Temporal visualizations saved.')
```

## 4. Statistical Visualization

```python
from src.visualization.statistical_viz import StatisticalVisualizer

viz = StatisticalVisualizer(output_dir='outputs/figures/statistical')
viz.plot_rot_distribution(df)
viz.plot_destination_change_rate(df)
viz.plot_correlation_heatmap(df)
viz.plot_feature_importance(df)
print('Statistical visualizations saved.')
```

## 5. Custom Plot Example

```python
# Speed distribution by vessel type
plt.figure(figsize=(12, 6))
df_sample = df.dropna(subset=['sog', 'ship_type_code']).sample(5000)
sns.boxplot(data=df_sample, x='ship_type_code', y='sog')
plt.xlabel('Ship Type Code')
plt.ylabel('Speed (knots)')
plt.title('Speed Distribution by Ship Type')
plt.tight_layout()
plt.show()
```
