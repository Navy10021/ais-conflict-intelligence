# EDA (Exploratory Data Analysis)

Explore the basic structure, missing values, and distribution of AIS data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.rcParams.update({'figure.figsize': (12, 6), 'font.size': 12})
%matplotlib inline
```

## 1. 데이터 로드

```python
# loader를 사용하여 데이터 로드
from src.preprocessing.loader import AISDKLoader

loader = AISDKLoader(data_root='./data', date_start='2024-03-01', date_end='2024-08-31')
df = loader.load()
print(f"총 레코드: {len(df):,}")
print(f"컬럼: {list(df.columns)}")
```

## 3. 결측값 분석

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_df[missing_df['missing_count'] > 0].sort_values('missing_count', ascending=False)
```

## 4. 선박 유형 분포

```python
plt.figure(figsize=(10, 6))
df['ship_type'].value_counts().head(15).plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.title('Top 15 Ship Types')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## 5. 시간별 데이터 분포

```python
df['date'] = df['timestamp'].dt.date
daily_counts = df.groupby('date')['mmsi'].nunique().reset_index()

plt.figure(figsize=(14, 6))
plt.plot(daily_counts['date'], daily_counts['mmsi'])
plt.xlabel('Date')
plt.ylabel('Unique Vessels')
plt.title('Daily Unique Vessels')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 6. 좌표 분포 (공간 시각화)

```python
plt.figure(figsize=(12, 8))
sample = df.dropna(subset=['lat', 'lon']).sample(min(50000, len(df)))
plt.scatter(sample['lon'], sample['lat'], alpha=0.1, s=1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('AIS Data Coverage (Sample)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. 속도(SOG) 분포

```python
plt.figure(figsize=(12, 6))
df['sog'].dropna().hist(bins=50, alpha=0.7)
plt.xlabel('Speed Over Ground (knots)')
plt.ylabel('Frequency')
plt.title('SOG Distribution')
plt.xlim(0, 30)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 8. 선박 상태(Navigational Status) 분포

```python
plt.figure(figsize=(10, 6))
df['nav_status'].value_counts().head(10).plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Navigational Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
